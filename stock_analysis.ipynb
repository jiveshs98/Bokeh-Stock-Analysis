```python
# Stock Analysis of Google
# Importing Libraries
# Download dataset with Python
from pandas_datareader import data

from datetime import datetime
from bokeh.plotting import figure, show, output_file

# Creating the Data object

start= datetime(2015,11,1)
end= datetime(2016,3,10)

df= data.DataReader(name="GOOG", data_source="yahoo", start= start, end= end)    # Data object is of dataframe type


def inc_dec(c, o):
    if c>o:
        value= "Increase"
    elif c<o:
        value= "Decrease"
    else:
        value= "Equal"
        
    return value

df["Status"]= [inc_dec(c,o) for c,o in zip(df.Close, df.Open)]


df["Middle"]= (df.Close+df.Open)/2
df["Height"]= abs(df.Close-df.Open)



# Creating the figure object
p= figure(x_axis_type='datetime', width=1000, height= 300, sizing_mode="scale_width")
p.title.text= "Candlestick Chart"

p.grid.grid_line_alpha=0.4  # Transparency of grid lines

hours_12= 12*60*60*1000   # Converting 12 hours into milliseconds for width of the rectangle



# Creating the segments
p.segment(df.index, df.High, df.index, df.Low, color="black")



# Creating the rectangles
# p.rect(x-coordinate of centre of rectangle, y coordinate of rectangle, width of rectangle, height of rectangle)
# When stock prices increased
p.rect(df.index[df.Status=="Increase"], df.Middle[df.Status=="Increase"], hours_12, df.Height[df.Status=="Increase"], 
       fill_color="#CCFFFF", line_color="black")



# When stock prices decreased
p.rect(df.index[df.Status=="Decrease"], df.Middle[df.Status=="Decrease"], hours_12, df.Height[df.Status=="Decrease"], 
       fill_color="#FF3333", line_color="black")




output_file("Candle.html")
show(p)
```
